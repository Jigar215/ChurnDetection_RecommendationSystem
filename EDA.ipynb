```python
# EDA for Churn Dataset
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load dataset
df = pd.read_csv("customer_churn_data.csv")
```

```python
# 1. Basic Info
print("Dataset Shape:", df.shape)
print("\nData Types:\n", df.dtypes)
print("\nMissing Values:\n", df.isnull().sum())
print("\nChurn Distribution:\n", df['Churn'].value_counts())
```

```python
# 2. Class Distribution Plot
plt.figure(figsize=(6,4))
sns.countplot(x='Churn', data=df, palette='Set2')
plt.title("Churn Distribution")
plt.show()
```

```python
# 3. Numerical Features Distribution
num_cols = ['Age', 'Tenure', 'MonthlyCharges', 'TotalCharges']
df[num_cols].hist(figsize=(12,8), bins=20, color='skyblue', edgecolor='black')
plt.suptitle("Numerical Features Distribution")
plt.show()
```

```python
# 4. Categorical Features vs Churn
cat_cols = ['Gender', 'ContractType', 'InternetService', 'TechSupport']

for col in cat_cols:
    plt.figure(figsize=(6,4))
    sns.countplot(x=col, hue='Churn', data=df, palette='Set1')
    plt.title(f"{col} vs Churn")
    plt.xticks(rotation=30)
    plt.show()
```

```python
# 5. Correlation Heatmap
plt.figure(figsize=(8,6))
corr = df[num_cols].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Heatmap (Numerical Features)")
plt.show()
```

```python
# 6. Boxplots (Churn impact on numeric features)
for col in num_cols:
    plt.figure(figsize=(6,4))
    sns.boxplot(x='Churn', y=col, data=df, palette='Set3')
    plt.title(f"{col} vs Churn")
    plt.show()
```
